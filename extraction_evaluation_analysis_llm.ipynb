```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Convert the averaged_metrics dictionary to a DataFrame for easier plotting
metrics_df = pd.DataFrame(averaged_metrics, index=[0])  # Use index=[0] for a single row DataFrame

# Set the plotting style
sns.set(style="whitegrid")

# 1. Bar Chart for Average Scores by Section
plt.figure(figsize=(12, 6))
score_columns = ['Endpoints_Score', 'Statistical_Score', 'Adverse_Events_Score', 'Clinical_Lab_Score', 'Total_Score']
sns.barplot(x=score_columns, y=metrics_df.loc[0, score_columns])
plt.title('Average Scores by Section')
plt.ylabel('Average Score')
plt.xlabel('Sections')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()


# 2. Bar Chart for Free Text Metrics
free_text_metrics = [
    'Endpoints_FreeText_Precision', 'Endpoints_FreeText_Recall', 'Endpoints_FreeText_F1',
    'Statistical_FreeText_Precision', 'Statistical_FreeText_Recall', 'Statistical_FreeText_F1',
    'Adverse_Events_FreeText_Precision', 'Adverse_Events_FreeText_Recall', 'Adverse_Events_FreeText_F1',
    'Clinical_Lab_FreeText_Precision', 'Clinical_Lab_FreeText_Recall', 'Clinical_Lab_FreeText_F1'
]

# Melting the DataFrame for easier plotting
free_text_df = metrics_df[free_text_metrics].mean().reset_index()
free_text_df.columns = ['Metric', 'Score']

plt.figure(figsize=(12, 6))
sns.barplot(data=free_text_df, x='Metric', y='Score')
plt.title('Average Free Text Metrics')
plt.ylabel('Average Score')
plt.xlabel('Metrics')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# 3. Box Plot for Score Distributions
plt.figure(figsize=(12, 6))
sns.boxplot(data=metrics_df[score_columns])
plt.title('Score Distribution Across Sections')
plt.ylabel('Scores')
plt.xlabel('Sections')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()



# Radar Chart for the first protocol's metrics (as an example)
from math import pi

def plot_radar_chart(data):
    labels = data.index.tolist()  
    num_vars = len(labels) 

    # Compute angle for each axis
    angles = [n / float(num_vars) * 2 * pi for n in range(num_vars)]
    data = data.values.flatten().tolist() 
    data += data[:1]  # Complete the loop
    angles += angles[:1] 

    plt.figure(figsize=(6, 6))
    ax = plt.subplot(111, polar=True)
    plt.xticks(angles[:-1], labels, color='grey', size=10)
    ax.plot(angles, data, linewidth=2, linestyle='solid', label='Average Metrics')
    ax.fill(angles, data, 'b', alpha=0.1)

    plt.title('Radar Chart for Average Metrics')
    plt.legend(loc='upper right')
    plt.show()
plot_radar_chart(metrics_df.iloc[0])  
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "llm_extraction_evaluation.xlsx"
protocol_sheets = [f'Sheet{i}' for i in range(1, 26)]
gold_standard_df = pd.read_excel(file_path, sheet_name='Gold standard')

protocol_data = {
    "Sheet": protocol_sheets,
    "Endpoints_Score": [],
    "Statistical_Score": [],
    "Adverse_Events_Score": [],
    "Clinical_Lab_Score": [],
    "Total_Score": [],
}

for sheet in protocol_sheets:
    df = pd.read_excel(file_path, sheet_name=sheet, header=None)
    protocol_data["Endpoints_Score"].append(df.iloc[15, 1])
    protocol_data["Statistical_Score"].append(df.iloc[16, 1])
    protocol_data["Adverse_Events_Score"].append(df.iloc[17, 1])
    protocol_data["Clinical_Lab_Score"].append(df.iloc[18, 1])
    protocol_data["Total_Score"].append(df.iloc[19, 1])

scores_df = pd.DataFrame(protocol_data)

merged_df = scores_df.merge(gold_standard_df, left_on="Sheet", right_on="Sheet", how="inner")

# Box Plot: Distribution of Total Scores by Phase
plt.figure(figsize=(10, 6))
sns.boxplot(data=merged_df, x='Phase', y='Total_Score')
plt.title('Distribution of Total Scores by Clinical Trial Phase')
plt.ylabel('Total Score')
plt.xlabel('Clinical Trial Phase')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()



# Bar Plot: Average Total Scores by Disease
plt.figure(figsize=(12, 6))
sns.boxplot(data=merged_df, x='Disease', y='Total_Score')
plt.title('Distribution of Total Scores by Disease')
plt.ylabel('Total Score')
plt.xlabel('Disease')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Grouping diseases into categories 

def disease_group(disease):
    d = str(disease).lower()
    if d in ["diabetes", "obesity, diabetes"]:
        return "Diabetes"
    elif d == "obesity":
        return "Obesity"
    elif d in [
        "cardiovascular disease",
        "non-alcoholic steatohepatitis (nash)",
        "liver disease",
        "hereditary haemochromatosis (hh)"
    ]:
        return "Cardiometabolic"
    else:
        return "Rare/Others"

merged_df['Disease_Group'] = merged_df['Disease'].apply(disease_group)

group_order = ["Diabetes", "Obesity", "Cardiometabolic", "Rare/Others"]

plt.figure(figsize=(12, 6))
sns.boxplot(
    data=merged_df,
    x='Disease_Group',
    y='Total_Score',
    order=group_order
)
plt.title('Distribution of Total Scores by Disease Group')
plt.xlabel('Disease Group')
plt.ylabel('Total Score')
plt.tight_layout()
plt.show()


plt.figure(figsize=(8, 5))
sns.barplot(
    data=merged_df,
    x='Disease_Group',
    y='Total_Score',
    order=group_order,
    estimator='mean',
    ci=None
)
plt.title('Mean Total Score by Disease Group')
plt.xlabel('Disease Group')
plt.ylabel('Mean Total Score')
plt.tight_layout()
plt.show()

```
